```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Gerar dados usando NumPy
x = np.linspace(0, 10, 100)  # 100 valores entre 0 e 10
y = np.sin(x)  # Função senoidal

```

```python
# Plotar os dados usando Matplotlib
plt.figure(figsize=(8, 4))
plt.plot(x, y, label='Seno de x', color='C1', linewidth=2)
plt.title('Gráfico de Função Senoidal')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.legend()
plt.show()
```
