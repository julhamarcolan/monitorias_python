```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parâmetros
z0 = 100  # posição inicial em metros
v0 = 5    # velocidade inicial em m/s
g = 9.8   # aceleração da gravidade em m/s^2
t_max = 10 # tempo máximo de simulação em segundos
```

```python
# Gerar pontos no tempo
t = np.linspace(0, t_max, num=500)
# Calcular a posição ao longo do tempo para dois objetos
z = z0 + v0 * t - 0.5 * g * t**2
```

```python
# Plotar o resultado
plt.plot(t, z, lw=2, color='pink')
plt.xlabel('Tempo (s)')
plt.ylabel('Altura (m)')
plt.title('Queda livre sem resistência do ar')
plt.grid(True)
plt.show()
```
