```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation as animation

#precisa ser usada apenas no google colab
from IPython.display import HTML
from matplotlib import rc
```

fig, ax = plt.subplots() cria uma figura (fig) e um conjunto de eixos (ax).

ax.set_xlim([0, 10]) define o limite do eixo x de 0 a 10.

```python
fig, ax = plt.subplots()
ax.set_xlim([0, 10])
```

scat = ax.scatter(1, 0) cria um ponto de dispersão inicial na posição (1, 0). Este ponto será movido ao longo do eixo x na animação.

x = np.linspace(0, 10) cria uma sequência de valores de 0 a 10 (usado para a posição x do ponto em cada frame da animação).

```python
scat = ax.scatter(1, 0)
x = np.linspace(0, 10)
```

A função animate(i) é chamada em cada frame da animação.

scat.set_offsets((x[i], 0)) atualiza a posição do ponto para (x[i], 0), movendo-o ao longo do eixo x.

return scat, retorna o objeto do ponto para atualizar a animação.

OBS: DIFERENÇA ENTRE O set_offsets e o set_data

set_offsets: Usado com objetos de dispersão, como aqueles criados por ax.scatter().

set_data: usado para atualizar coordenadas em gráficos de linha, onde x e y são fornecidos como listas ou arrays separados para a linha traçada (criados com ax.plot).



```python
def animate(i):
    scat.set_offsets((x[i], 0))
    return scat,
```

```python
ani = animation.FuncAnimation(fig, animate, repeat=True,
                              frames=len(x) - 1, interval=50)
```
